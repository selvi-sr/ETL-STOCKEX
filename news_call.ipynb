```python
import requests
import json
import pandas as pd
import math
import pprint
```

```python
api_key = 'apiKey=2d229173849a4caa959cc064f8906dd0'
```

```python
base_url = 'https://newsapi.org/v2/everything?'
```

```python
from_date = '2019-10-15'
from_search = f'&from={from_date}'

to_date = '2019-10-31'
to_search = f'&to={to_date}'
```

```python
page = '1'
search_page = f'&page={page}'
```

```python
tickers = ["AMZN","CSCO","MSFT", "NFLX","GOOGL"]
```

```python
publish_dates = []
article_titles = []
Company_Id = []
for ticker in tickers:
    search_ticker = f'&q={ticker}'
    requests.get(f"{base_url}{api_key}{search_ticker}{from_search}{to_search}{search_page}").json()['totalResults']
    for page in range(1,6):
        search_page = f'&page={str(page)}'
        myresults = requests.get(f"{base_url}{api_key}{search_ticker}{from_search}{to_search}{search_page}").json()
        for article in myresults['articles']:
            Company_Id.append(tickers.index(ticker)+1)
            publish_date = article['publishedAt']
            publish_date = publish_date.split("T")[0]
            publish_dates.append(pd.to_datetime(publish_date))
            article_titles.append(article['title'])
        
my_company = pd.DataFrame({
    'Company_Id': Company_Id,
    'title': article_titles,
    'publish date': publish_dates
})   
```
